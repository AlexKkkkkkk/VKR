# Парсинг студий и 1-комн.квартир, новостройки, не апартаменты

```python
data = []

for p in range(1, 55):
    print(p)
    url = (f'https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&foot_min=45&object_type%5B0%5D=2&offer_type=flat&only_flat=1&only_foot=2&p={p}&region=1&room1=1&room9=1')
    r = requests.get(url)
    sleep(1)
    soup = BeautifulSoup(r.text, 'lxml')
    flats = soup.findAll('article', class_='_93444fe79c--container--Povoi _93444fe79c--cont--OzgVc')
    for flat in flats:
        try:
            link = flat.find('a', class_ = '_93444fe79c--link--eoxce').get('href')
        except AttributeError:
             continue
        data.append([link])
```
